Question 1. What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.


- Anomaly detection is the process of identifying data points, events, or observations that deviate significantly from the expected or normal pattern of a dataset. There are three main types: Point Anomalies, which are single, unusual data points; Contextual Anomalies, which are data points anomalous only within a specific context; and Collective Anomalies, which are a collection of related data points that together show abnormal behavior, even if individual points appear normal.
- Point Anomaly (Global Outlier)
Explanation: A single data instance that is significantly different from the rest of the dataset.
Example: A sudden, unusually large transaction on a credit card, far exceeding the customer's typical spending.
- Contextual Anomaly
Explanation: A data point that is abnormal only in a specific context but might be considered normal in a different situation.
Example: High energy consumption at 3 AM in a residential home is a contextual anomaly, as this level of usage is normal during the day but unusual late at night.
- Collective Anomaly
Explanation: A group of related data points that, when considered together over time or space, exhibit an abnormal pattern, even if individual points don't seem unusual.
Example: A sequence of slightly irregular heartbeats on an ECG, even though each individual beat might look normal, could signal an anomaly in heart function.

Question 2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases.

- Isolation Forest identifies outliers by randomly partitioning data, LOF detects anomalies by comparing a point's local density to its neighbors, and DBSCAN groups dense points into clusters and labels sparse points as noise. Isolation Forest excels at large datasets and global outliers but struggles with local anomalies, while LOF is excellent for detecting local anomalies but is computationally expensive. DBSCAN identifies clusters and noise based on density, making it suitable for data with complex shapes and for identifying outliers that are simply not part of any dense group.
- Isolation Forest (iForest)
Approach: An ensemble method that builds random trees. Outliers are points that are isolated in fewer partitions (shorter paths) in these trees.
Suitable Use Cases:
Large datasets, as it is fast and scalable.
Identifying global outliers, but less effective for local anomalies.
When you need a method that requires less parameter tuning.
- Local Outlier Factor (LOF)
Approach: A density-based algorithm that measures the local density of a point relative to its neighbors. Points with significantly lower local density are outliers.
Suitable Use Cases:
Detecting local outliers.
High-dimensional data.
Datasets where anomalies are defined by their lack of proximity to other data points.
- DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
Approach: Groups together points that are closely packed together in feature space, forming dense regions, and labels points in sparsely populated regions as noise (outliers).
Suitable Use Cases:
Discovering clusters of arbitrary shapes.
Identifying outliers that are simply not part of any dense cluster.
Noisy data, as it can handle noise and doesn't require pre-defining the number of clusters.

Question 3. What are the key components of a Time Series? Explain each with one
example.


- The four key components of a time series are Trend, Seasonality, Cyclicality, and Irregularity. A Trend is the long-term direction of data over time, while Seasonality represents predictable, shorter-term patterns that repeat within a fixed, short interval, like a year. Cyclicality refers to fluctuations that occur over longer, non-fixed periods, and Irregularity accounts for random, unpredictable variations or "noise" in the data.

Question 4. : Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?

- A time series is stationary if its statistical properties, such as mean, variance, and autocorrelation, remain constant over time, indicating a lack of trend or seasonality. You can test for stationarity using visual methods like run sequence plots or statistical tests such as the Augmented Dickey-Fuller (ADF) test, which looks for a unit root. To make a non-stationary series stationary, you can apply transformations like differencing to remove trends and seasonality, or use logarithmic transformations to stabilize variance.  
- Testing for Stationarity
Visual Inspection: Plot the time series to visually check for trends, seasonality, or changes in variance.
Summary Statistics: Compare the mean and variance of different time partitions. A significant difference suggests non-stationarity.
Statistical Tests:
Augmented Dickey-Fuller (ADF) Test: This test checks for the presence of a unit root, a common cause of non-stationarity. A low p-value (typically < 0.05) leads to rejection of the null hypothesis (the series is stationary), suggesting the series is non-stationary.
Kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test: The null hypothesis of the KPSS test is that the series is stationary.

Question 5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.


- The AR, MA, ARIMA, SARIMA, and SARIMAX models are all used for time series forecasting, with increasing complexity and capability to handle different data characteristics.
1. AR (Autoregressive) Model:
Structure: Predicts future values based on a linear combination of past values of the series itself. Denoted as AR(p), where 'p' is the order of the autoregressive component (number of past values used).
Application: Suitable for stationary time series where the current value is primarily dependent on its recent past.
2. MA (Moving Average) Model:
Structure: Predicts future values based on a linear combination of past forecast errors (residuals). Denoted as MA(q), where 'q' is the order of the moving average component (number of past error terms used).
Application: Suitable for stationary time series where the current value is primarily influenced by past random shocks or innovations.
3. ARIMA (Autoregressive Integrated Moving Average) Model:
Structure: Combines AR and MA components with an "Integrated" (I) component. Denoted as ARIMA(p, d, q), where 'd' is the order of differencing required to make the series stationary. Differencing removes trends and seasonality.
Application: Used for non-stationary time series that can be made stationary through differencing. It handles trends but not explicitly seasonality.
4. SARIMA (Seasonal Autoregressive Integrated Moving Average) Model:
Structure: Extends ARIMA by adding seasonal components. Denoted as SARIMA(p, d, q)(P, D, Q)s, where (p, d, q) are the non-seasonal orders, (P, D, Q) are the seasonal orders, and 's' is the length of the seasonal period.
Application: Specifically designed for time series exhibiting clear seasonal patterns in addition to trends.
5. SARIMAX (Seasonal Autoregressive Integrated Moving Average with Exogenous Regressors) Model:
Structure: Extends SARIMA by incorporating exogenous (external) variables. It includes all the components of SARIMA and adds a linear regression component for the exogenous variables.
Application: Used when the time series is influenced by external factors (e.g., economic indicators, weather data) that can be included as predictors alongside the time series' own past values and errors. It handles trends, seasonality, and external influences.

Question 8. Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.


```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

data["Month"] = pd.to_datetime(data["Month"])
data.set_index("Month", inplace=True)
data.rename(columns={"#Passengers": "Passengers"}, inplace=True)
data.index.freq = "MS"  # Set the frequency to Monthly Start

model = SARIMAX(
    data["Passengers"],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False,
)
results = model.fit()

# Step 4: Forecast the next 12 months
forecast_steps = 12
forecast = results.get_forecast(steps=forecast_steps)
forecast_df = forecast.summary_frame()

# Step 5: Visualize the results
plt.figure(figsize=(12, 6))
plt.plot(data["Passengers"], label="Observed")
plt.plot(forecast_df["mean"], label="Forecast", color="red")
plt.fill_between(
    forecast_df.index,
    forecast_df["mean_ci_lower"],
    forecast_df["mean_ci_upper"],
    color="pink",
    alpha=0.3,
)
plt.title("Monthly Airline Passengers: SARIMA Forecast")
plt.xlabel("Date")
plt.ylabel("Number of Passengers")
plt.legend()
plt.grid(True)
plt.show()

```

Question 9. Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
np.random.seed(42)
# Generate inliers (normal data points)
X_inliers = 0.3 * np.random.randn(100, 2)
X_inliers = np.r_[X_inliers + 2, X_inliers - 2] # Create two clusters

# Generate outliers (anomalies)
X_outliers = np.random.uniform(low=-6, high=6, size=(20, 2))
X = np.r_[X_inliers, X_outliers] # Combine inliers and outliers
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = lof.fit_predict(X) # Predict inliers (1) and outliers (-1)
X_scores = lof.negative_outlier_factor_
plt.figure(figsize=(10, 7))

# Plot inliers
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], label='Inliers', c='blue', s=50)

# Plot outliers
plt.scatter(X[y_pred == -1, 0], X[y_pred == -1, 1], label='Outliers', c='red', s=100, marker='x')

plt.title('Anomaly Detection using Local Outlier Factor (LOF)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

print("Predicted labels (1 for inlier, -1 for outlier):")
print(y_pred)
print("\nNegative LOF scores (lower values indicate higher outlierness):")
print(X_scores)
```
